```python
# --- Configuration ---
S3_BUCKET = "s3://tsnpdcl-analytics-datalake-poc-local"
BRONZE_S3_PATH = f"{S3_BUCKET}/bronze/consumption"
SILVER_S3_PATH = f"{S3_BUCKET}/silver/consumption"
QUARANTINE_PATH = f"{S3_BUCKET}/data/quarantine/silver/"
CHECKPOINT_PATH = f"{S3_BUCKET}/checkpoints/silver"
```

```python
# --- Imports ---
import traceback
import logging
import sys
from pyspark.sql.functions import upper, when, col, current_timestamp
from delta.tables import DeltaTable
from py4j.protocol import Py4JError
```

```python
# Configure root logger to stream to stdout so Databricks Notebooks show INFO logs
logging.basicConfig(level=logging.INFO, 
                    stream=sys.stdout,
                    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
                    force=True)
logger = logging.getLogger("SilverETL")
```

```python
def process_microbatch(df, batch_id):
    """Processes a single micro-batch of data incrementally from Bronze to Silver."""
    initial_count = df.count()
    if initial_count == 0:
        logger.info(f"Batch {batch_id} is empty. Skipping processing.")
        return

    logger.info(f"Processing Batch {batch_id} with {initial_count} rows from Bronze.")

    # --- 1. Data Quality & Quarantine ---
    # Production best practice: Don't just filter bad records out; save them to analyze later.
    bad_records_df = df.filter(col("extraction_year").isNull() | 
                               col("extraction_month").isNull() | 
                               col("Units").isNull() | 
                               col("Load").isNull() | 
                               col("Section").isNull())
    
    # Save bad records to quarantine
    bad_count = bad_records_df.count()
    if bad_count > 0:
        logger.warning(f"Found {bad_count} invalid records in Batch {batch_id}. Sending to Quarantine.")
        bad_records_df.withColumn("quarantined_at", current_timestamp()) \
            .write.format("delta").mode("append").save(QUARANTINE_PATH)

    # Keep only valid records
    valid_df = df.filter(col("extraction_year").isNotNull() & 
                         col("extraction_month").isNotNull() & 
                         col("Units").isNotNull() & 
                         col("Load").isNotNull() & 
                         col("Section").isNotNull())

    # --- 2. Standardization ---
    try:
        standardized_df = valid_df \
            .withColumn("Circle", upper(col("Circle"))) \
            .withColumn("Division", upper(col("Division"))) \
            .withColumn("SubDivision", upper(col("SubDivision"))) \
            .withColumn("Section", upper(col("Section"))) \
            .withColumn("Area", upper(col("Area"))) \
            .withColumn("Units", col("Units").cast("double")) \
            .withColumn("Load", col("Load").cast("double")) \
            .withColumn("Season", 
                when(col("extraction_month").between(4, 6), "Summer")
                .when((col("extraction_month") >= 11) | (col("extraction_month") == 1), "Winter")
                .otherwise("Monsoon")) \
            .withColumn("Financial_Quarter",
                when(col("extraction_month").between(4, 6), "Q1")
                .when(col("extraction_month").between(7, 9), "Q2")
                .when(col("extraction_month").between(10, 12), "Q3")
                .otherwise("Q4"))
    except Exception as transform_e:
        logger.error(f"Error during standardization in Batch {batch_id}: {transform_e}")
        raise transform_e

    # --- 3. Deduplication ---
    deduplicated_df = standardized_df.dropDuplicates([
        'Circle', 'Division', 'SubDivision', 'Section', 'Area', "extraction_month", "extraction_year"
    ])
    
    final_count = deduplicated_df.count()
    logger.info(f"Deduplication complete. Dropped {initial_count - bad_count - final_count} duplicates.")

    # --- 4. Incremental Load (Upsert / Merge) to Silver ---
    if final_count > 0:
        logger.info(f"Merging {final_count} cleaned records into Silver path.")
        
        # Check if Silver table exists by checking if it's a DeltaTable
        if DeltaTable.isDeltaTable(spark, SILVER_S3_PATH):
            silver_table = DeltaTable.forPath(spark, SILVER_S3_PATH)
            
            # Perform atomic MERGE
            silver_table.alias("target").merge(
                deduplicated_df.alias("source"),
                condition="""
                    target.Circle = source.Circle AND
                    target.Division = source.Division AND
                    target.SubDivision = source.SubDivision AND
                    target.Section = source.Section AND
                    target.Area = source.Area AND
                    target.extraction_month = source.extraction_month AND
                    target.extraction_year = source.extraction_year
                """
            ).whenMatchedUpdateAll() \
             .whenNotMatchedInsertAll() \
             .execute()
            logger.info(f"Merge operation completed for Batch {batch_id}.")
        else:
            # First run: Initialize the Delta table
            logger.info("Silver table does not exist. Initializing with Batch 0.")
            deduplicated_df.write \
                .format("delta") \
                .mode("overwrite") \
                .partitionBy("extraction_year") \
                .save(SILVER_S3_PATH)
```

```python

try:
    logger.info("Starting Incremental Silver Layer Streaming Job")
    
    # 1. Read Stream from Bronze Delta Table
    bronze_stream = spark.readStream.format("delta").load(BRONZE_S3_PATH)
    
    # 2. Process stream incrementally in micro-batches
    query = bronze_stream.writeStream \
        .foreachBatch(process_microbatch) \
        .option("checkpointLocation", CHECKPOINT_PATH) \
        .trigger(availableNow=True) \
        .start()
        
    query.awaitTermination()
    logger.info("Silver Incremental Load completed successfully.")

except Exception as e:
    logger.error(f"FATAL ERROR in Silver Stream: {e}")
    logger.error(traceback.format_exc())
    raise e
```
